## Draw distribution plot

```python
from pyecharts import options as opts
from pyecharts.charts import Page, Pie
```

```python
df['Case:'].unique()
case_df=df['Case:']
case_df=case_df.value_counts(ascending=True, dropna=False)
case_dict = dict(case_df)
name=pd.DataFrame(case_dict.keys())
value=pd.DataFrame(case_dict.values())
name=name[0].tolist()
value=value[0].tolist()
words=list(zip(list(name),list(value)))
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Page, Pie
h=list(zip( list(case_dict.keys()),list(case_dict.values())))
pie = Pie()
pie.add("",words)
pie.set_global_opts(title_opts=opts.TitleOpts(title="Pie Chart of Number of Cases distributed"), 
                    legend_opts=opts.LegendOpts(pos_left="left", orient="vertical", pos_top='10%'))
pie.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
# pie.load_javascript() # Uncomment for Jupyter notebook
```

```python
pie.render_notebook()
```

```python
df['Metric:'].unique()
model_df=df['Metric:']
model_df=model_df.value_counts(ascending=True, dropna=False)
model_dict = dict(model_df)
name=pd.DataFrame(model_dict.keys())
value=pd.DataFrame(model_dict.values())
name=name[0].tolist()
value=value[0].tolist()
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Page, Pie, Bar
words=list(zip(list(name),list(value)))
words.sort(key=lambda x: x[1])
(
    Pie(init_opts=opts.InitOpts(bg_color="#2c343c")) # bg_color="#2c343c"
    .add(
        series_name="name of Algorithm",
        data_pair=words,
        rosetype="radius",
        radius="55%",
        center=["50%", "50%"],
        label_opts=opts.LabelOpts(is_show=False, position="center"),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="Pie Chart of Number of metric distributed",
            pos_left="center",
            pos_top="20",
            title_textstyle_opts=opts.TextStyleOpts(color="#fff"),
        ),
        legend_opts=opts.LegendOpts(is_show=False),
    )
    .set_series_opts(
        tooltip_opts=opts.TooltipOpts(
            trigger="item", formatter="{a} <br/>{b}: {c} ({d}%)"
        ),
        label_opts=opts.LabelOpts(color="rgba(255, 255, 255, 0.3)"),
    )
    .render_notebook()
)
```

```python
df['Algorithm'].unique()
model_df=df['Algorithm']
model_df=model_df.value_counts(ascending=True, dropna=False)
model_dict = dict(model_df)
name=pd.DataFrame(model_dict.keys())
value=pd.DataFrame(model_dict.values())
name=name[0].tolist()
value=value[0].tolist()
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Page, Pie, Bar

c = (
    Pie()
    .add(
        "",
        [list(z) for z in (zip(list(name),list(value)))],
        radius=["40%", "55%"],
        label_opts=opts.LabelOpts(
            position="outside",
            formatter="{a|{a}}{abg|}\n{hr|}\n {b|{b}: }{c}  {per|{d}%}  ",
            background_color="#eee",
            border_color="#aaa",
            border_width=1,
            border_radius=4,
            rich={
                "a": {"color": "#999", "lineHeight": 22, "align": "center"},
                "abg": {
                    "backgroundColor": "#e3e3e3",
                    "width": "100%",
                    "align": "right",
                    "height": 22,
                    "borderRadius": [4, 4, 0, 0],
                },
                "hr": {
                    "borderColor": "#aaa",
                    "width": "100%",
                    "borderWidth": 0.5,
                    "height": 0,
                },
                "b": {"fontSize": 16, "lineHeight": 33},
                "per": {
                    "color": "#eee",
                    "backgroundColor": "#334455",
                    "padding": [2, 4],
                    "borderRadius": 2,
                },
            },
        ),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="Pie Chart of Number of Algorithm distributed"),legend_opts=opts.LegendOpts(is_show=False))
)
c.render_notebook()
```
