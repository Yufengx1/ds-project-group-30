```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Cleansing the data/result clean data.xlsx")
df
```

```python
df.loc[df["Gantry:"] == "combined", "Gantry:"] = 1
```

```python
df.head()
```

```python
Manuf = df.groupby(['Manuf'])['predictedValue'].mean().reset_index(name="mean")
Manuf
```

```python
Model = df.groupby(['Model'])['predictedValue'].mean().reset_index(name="mean")
Model
```

```python
TPS = df.groupby(['TPS'])['predictedValue'].mean().reset_index(name="mean")
TPS
```

```python
Algorithm = df.groupby(['Algorithm'])['predictedValue'].mean().reset_index(name="mean")
Algorithm
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18,15))
Manuf[["Manuf","mean"]].plot(ax=axes[0][0], kind='bar', xlabel="Manufacturer")
Model[["Model", "mean"]].plot(ax=axes[0][1], kind='bar', xlabel="Model")
TPS [["TPS","mean"]].plot(ax=axes[1][0], kind='bar', xlabel="TPS")
Algorithm[["Algorithm", "mean"]].plot(ax=axes[1][1], kind='bar', xlabel="Algorithm")
```
